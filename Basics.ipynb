# Visualizing Earnings Based On College Majors #

In this guided project, we'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

* __Rank__ - Rank by median earnings (the dataset is ordered by this column).
* __Major_code__ - Major code.
* __Major__  - Major description.
* __Major_category__ - Category of major.
* __Total__ - Total number of people with major.
* __Sample_size__ - Sample size (unweighted) of full-time.
* __Men__ - Male graduates.
* __Women__ - Female graduates.
* __ShareWomen__ - Women as share of total.
* __Employed__ - Number employed.
* __Median__ - Median salary of full-time, year-round workers.
* __Low_wage_jobs__ - Number in low-wage service jobs.
* __Full_time__ - Number employed 35 hours or more.
* __Part_time__ - Number employed less than 35 hours.

Using visualizations, we can start to explore questions from the dataset like:

* Do students in more popular majors make more money?
* How many majors are predominantly male? Predominantly female?
* Which category of majors have the most students?

Now we'll  compare the **number of men with the number of women in each category of majors**.

```python

import numpy as np

part1 = recent_grads[:60]
part2 = recent_grads[60:121]
part3 = recent_grads[121:]

for i in range(6):
    start = 1 + i*10
    end = start + 10
    labels = part1['Major'][start:end]
    men = part1['Men'][start:end]
    women = part1['Women'][start:end]
    
    n = np.arange(len(labels))
    
    fig, ax = plt.subplots()
    ax.bar(x=n, height=men, width=0.35, label='Men', color='red')
    ax.bar(x=n+0.3, height=women, width=0.35, label='Women', color='blue')
    ax.set_xticks(n)
    ax.set_xticklabels(labels, rotation=90)
    plt.title('Men vs Women')
    plt.legend(loc='upper left')
```

```python
for i in range(6):
    start = 1 + i*10
    end = start + 10
    labels = part2['Major'][start:end]
    men = part2['Men'][start:end]
    women = part2['Women'][start:end]
    
    n = np.arange(len(labels))
    
    fig, ax = plt.subplots()
    ax.bar(x=n, height=men, width=0.35, label='Men', color='red')
    ax.bar(x=n+0.3, height=women, width=0.35, label='Women', color='blue')
    ax.set_xticks(n)
    ax.set_xticklabels(labels, rotation=90)
    plt.title('Men vs Women')
    plt.legend(loc='upper left')
```

```python
for i in range(5):
    start = 1 + i*10
    end = start + 10
    labels = part3['Major'][start:end]
    men = part3['Men'][start:end]
    women = part3['Women'][start:end]
    
    n = np.arange(len(labels))
    
    fig, ax = plt.subplots()
    ax.bar(x=n, height=men, width=0.35, label='Men', color='red')
    ax.bar(x=n+0.3, height=women, width=0.35, label='Women', color='blue')
    ax.set_xticks(n)
    ax.set_xticklabels(labels, rotation=90)
    plt.title('Men vs Women')
    plt.legend(loc='upper left')
```

Men dominate in the first 50 majors. Then the distribution evens out. And the last 50 majors are dominated by women.

Now we'll have a look at the distributions across different types of majors. We will analyse the following majors:

```python
engineering = recent_grads[recent_grads['Major'].str.contains('ENGINEERING')]

n=np.arange(len(engineering['Major']))
fig,ax = plt.subplots()
ax.bar(x=np.arange(len(engineering)), height=engineering['Women'], width=0.4, label='Women', color='red')
ax.bar(x=np.arange(len(engineering))+0.3, height=engineering['Men'], width=0.4, label='Men', color='blue')
ax.set_xticks(n)
ax.set_xticklabels(engineering['Major'],rotation=90)
plt.title('Engineering Gender Distribution')
plt.legend(loc='upper left')
```

Engineering majors are greatly dominated by men.

```python
science = recent_grads[recent_grads['Major'].str.contains('SCIENCE')]
s=np.arange(len(science['Major']))

fig,ax = plt.subplots()
ax.bar(x=s, height=science['Women'], width=0.4, label='Women', color='red')
ax.bar(x=s+0.3, height=science['Men'], width=0.4, label='Men', color='blue')
ax.set_xticks(s)
ax.set_xticklabels(science['Major'], rotation=90)
plt.title('Science Major Gender Distribution')
plt.legend(loc='upper left')
```

The situation is better with science majors. Here the amount of men and women is almost equal. Though the trend is the following: well-paid majors are dominated by men, ill-paid are dominated by women/.

```python
technologies = recent_grads[recent_grads['Major'].str.contains('TECHNOLOGIES')]
t = np.arange(len(technologies['Major']))

fig,ax = plt.subplots()
ax.bar(x=t, height=technologies['Women'], width=0.4, label='Women', color='red')
ax.bar(x=t+0.3, height=technologies['Men'], width=0.4, label='Men', color='blue')
plt.title('Technologies Major Gender Distribution')
ax.set_xticks(t)
ax.set_xticklabels(technologies['Major'], rotation=90)
plt.legend(loc='upper left')
```

Here the distribution is uneven, though we can say that men dominate 8 out of 10 technologies majors. 

```python
education = recent_grads[recent_grads['Major'].str.contains('STUDIES|EDUCATION')]
e=np.arange(len(education['Major']))
fig,ax = plt.subplots()
ax.bar(x=e, height=education['Women'], width=0.4, label='Women', color='red')
ax.bar(x=e+0.3, height=education['Men'], width=0.4, label='Men', color='blue')
ax.set_xticks(e)
ax.set_xticklabels(education['Major'], rotation=90)
plt.title('Education Major Gender Distribution')
plt.legend(loc='upper left')
```

The educational field is dominated by women everywhere except 'PHILOSOPHY AND RELIGIOUS STUDIES'. The most popular majors among women are 'ELEMENTARY EDUCATION' and 'GENERAL EDUCATION'. 

```python
arts = recent_grads[recent_grads['Major'].str.contains('ARTS')]
a = np.arange(len(arts['Major']))

fig,ax = plt.subplots()
ax.bar(x=a, height=arts['Women'], width=0.4, label='Women', color='red')
ax.bar(x=a+0.3, height=arts['Men'], width=0.4, label='Men', color='blue')
ax.set_xticks(a)
ax.set_xticklabels(arts['Major'],rotation=90)
plt.title('Arts Major Gender Distribution')
plt.legend(loc='upper left')
```

All the majors except 'MISCELLANEOUS FINE ARTS' and 'FILM VIDEO AND PHOTOGRAPHIC ARTS' are dominated by women.

### Conclusion###

Now we'll *explore the distributions of median salaries and unemployment rate*.

### The Distributions of Median Salaries and Unemployment Rate###

```python
fig,ax = plt.subplots()
ax.boxplot(recent_grads['Median'])
plt.title('Median Salaries')
```

We can see that the minimum median salary is a bit higher than 20.000. The maximum is a bit higher than 60.000. The median here is around 35.000. So half of all callege majors graduates earn not more than 35.000. And another half earns from 35.000 to 60.000. 

```python
fig,ax = plt.subplots()
ax.boxplot(recent_grads['Unemployment_rate'])
plt.title('Unemployment Rate')
```

The unemployment rate minimum is around zero and maximum is around 0.125. A half of all majors have from 0 to 0.07 unemployment rate. Another half has from 0.7 to 0.125.
